9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
import pandas as pd
date = datetime.datetime.strptime("01-1-2023", "%d-%m-%Y")
k=10
date_generated = pd.date_range(date,periods=k)
print(date_generated.strftime("%d-%m-%Y"))
```
